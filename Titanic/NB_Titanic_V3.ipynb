# Pre-processing

```python
# To save the score and the parameters in an external file for comparison
summary = pd.read_csv('results/0-summary.csv')
```

```python
from sklearn.model_selection import train_test_split

# Separate target from predictors
X = train.drop(columns=['PassengerId', 'Survived'])
y = train[['Survived']].copy()

# Divide data into training and validation subsets
X_train_all_columns, X_valid_all_columns, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn import set_config
set_config(transform_output="pandas")
```

```python
# Select categorical columns with relatively low cardinality (<10 --> convenient but arbitrary)
low_card_columns = [cname for cname in X_train_all_columns.columns if X_train_all_columns[cname].nunique() < 10 and 
                        X_train_all_columns[cname].dtype == "object"]

# Select numerical columns
numerical_cols = [cname for cname in X_train_all_columns.columns if X_train_all_columns[cname].dtype in ['int64', 'float64']]

# Keep selected columns only
my_cols = low_card_columns + numerical_cols
X_train = X_train_all_columns[my_cols].copy()
X_valid = X_valid_all_columns[my_cols].copy()
```

```python
# Preprocessing for numerical data
numerical_transformer = SimpleImputer(strategy='constant')

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, low_card_columns)
    ])
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train))
X_valid = pd.DataFrame(preprocessor.transform(X_valid))
```

```python
pd.DataFrame([[datetime.now(tz=None).isoformat(' ', "seconds"),str(preprocessor),"N/A","N/A"]], columns=summary.columns).to_csv("results/0-summary.csv", index=False, header=False, mode='a')
summary = pd.concat([summary, pd.DataFrame([[datetime.now(tz=None).isoformat(' ', "seconds"),str(preprocessor),"N/A","N/A"]], columns=summary.columns)])
```

## Neural Network

```python
from tensorflow import keras
from tensorflow.keras import layers # type: ignore
from tensorflow.keras.callbacks import EarlyStopping # type: ignore
```

```python
model_DNN = keras.Sequential()

model_DNN.add(keras.layers.Input(shape=[X_train.shape[1]]))
model_DNN.add(keras.layers.BatchNormalization(),)

# Hidden layers
layers = 6
act = 'relu'  # swish or relu
node = 256
dropout = 0.5

for i in range(layers):
    model_DNN.add(keras.layers.Dense(node, activation=act))
    model_DNN.add(keras.layers.Dropout(dropout))
    model_DNN.add(keras.layers.BatchNormalization())
    
model_DNN.add(keras.layers.Dense(y_train.shape[1]))
model_DNN.add(keras.layers.Activation('sigmoid'))

early_stopping = EarlyStopping(
    min_delta=0.1, # minimium amount of change to count as an improvement
    patience=30, # how many epochs to wait before stopping
    restore_best_weights=True,
)

# we'll use categorical xent for the loss, and RMSprop as the optimizer
model_DNN.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print("Training...")
history = model_DNN.fit(X_train, y_train, 
              validation_data=(X_valid, y_valid),
              epochs=100, 
              batch_size=50,
              callbacks=[early_stopping],
              verbose=0
)

history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(history_df['val_loss'].min()))

print("Evaluation: ")
print(model_DNN.evaluate(X_valid, y_valid))

print("Generating test predictions...")
y_predicted = (model_DNN.predict(X_valid, verbose=0)>= 0.5).astype(int)
print("Predictions generated")

print("The mean absolute error on the validation set is: ")
print(round(mean_absolute_error(y_valid['Survived'], y_predicted), 3))
print("The absolute score on the validation set is: ")
print(round(sklearn.metrics.accuracy_score(y_valid['Survived'], y_predicted, normalize=True), 3))
print("The normalized score on the validation set is: ")
print(sklearn.metrics.accuracy_score(y_valid['Survived'], y_predicted, normalize=False))

pd.DataFrame([[datetime.now(tz=None).isoformat(' ', "seconds"),"N/A",
               f"layers = {layers}, act = {act}, node = {node} dropout = {dropout}",
               str(round(sklearn.metrics.accuracy_score(y_valid['Survived'], y_predicted, normalize=True), 3))]], 
             columns=summary.columns).to_csv("results/0-summary.csv", index=False, header=False, mode='a')
summary = pd.concat([summary, pd.DataFrame([[datetime.now(tz=None).isoformat(' ', "seconds"),"N/A",
               f"layers = {layers}, act = {act}, node = {node} dropout = {dropout}",
               str(round(sklearn.metrics.accuracy_score(y_valid['Survived'], y_predicted, normalize=True), 3))]], 
             columns=summary.columns)])
```
